```python
import pandas as pd
import csv
```

```python
df_license_19 = pd.read_csv("./data/driver_license_19.csv", encoding="cp949")
```

```python
df_license_19.rename(columns={"구분": "age_group"}, inplace=True)
df_license_19 = df_license_19.set_index("age_group")
```

```python
df_license_19
```

```python
df_license_19["total"] = df_license_19.sum(axis=1)
df_license_19 = df_license_19[["total"]]
```

```python
df_license_19
```

```python
def change_index(index):
    if index == "99 이상":
        return 65
    if 16 <= int(index) <= 19:
        return 10
    elif 20 <= int(index) <= 29:
        return 20
    elif 30 <= int(index) <= 39:
        return 30
    elif 40 <= int(index) <= 49:
        return 40
    elif 50 <= int(index) <= 59:
        return 50
    elif 60 <= int(index) <= 64:
        return 60
    else:
        return 65
```

```python
df_license_19.index = df_license_19.index.map(change_index)
```

```python
df_license_19
```

```python
df_license_19.to_csv("./data/pre_driver_license_19.csv", encoding="utf-8")
```
