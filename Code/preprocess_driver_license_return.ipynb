```python
import pandas as pd
import numpy as np
import csv
```

```python
df_return = pd.read_csv("./data/driver_license_return.csv", encoding="cp949")
```

```python
df_return.rename(columns={"지방청": "city"}, inplace=True)
df_return["city"] = df_return["city"].replace({"경기남부": "경기", "경기북부": "경기"})
df_return = df_return.groupby("city").sum().reset_index()
df_return = df_return.set_index("city")
```

```python
df_return
```

```python
df_return = df_return[df_return.columns[df_return.columns.str.contains("2019|2020|2021|2022|2023")]]
df_return = df_return[df_return.columns[df_return.columns.str.contains("65")]]
df_return.columns = df_return.columns.str.replace(r'^(\d{4}) (\d{2}세 이상|65세이상)$', r'elderly_\1', regex=True)
```

```python
df_return
```

```python
df_return.to_csv("./data/pre_driver_license_return.csv", encoding="utf-8")
```
