```python
import pandas as pd
import csv
```

```python
df_license_20 = pd.read_csv("./data/driver_license_20.csv", encoding="cp949")
```

```python
df_license_20.rename(columns={"구분": "age_group"}, inplace=True)
df_license_20 = df_license_20.set_index("age_group")
```

```python
df_license_20
```

```python
df_license_20["total"] = df_license_20.sum(axis=1)
df_license_20 = df_license_20[["total"]]
```

```python
df_license_20
```

```python
def change_index(index):
    if index == "99 이상":
        return 65
    if 16 <= int(index) <= 19:
        return 10
    elif 20 <= int(index) <= 29:
        return 20
    elif 30 <= int(index) <= 39:
        return 30
    elif 40 <= int(index) <= 49:
        return 40
    elif 50 <= int(index) <= 59:
        return 50
    elif 60 <= int(index) <= 64:
        return 60
    else:
        return 65
```

```python
df_license_20.index = df_license_20.index.map(change_index)
```

```python
df_license_20
```

```python
df_license_20.to_csv("./data/pre_driver_license_20.csv", encoding="utf-8")
```
