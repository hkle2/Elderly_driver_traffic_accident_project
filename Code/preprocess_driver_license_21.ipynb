```python
import pandas as pd
import csv
```

```python
df_license_21 = pd.read_csv("./data/driver_license_21.csv", encoding="cp949")
```

```python
df_license_21.rename(columns={"연령대": "age_group"}, inplace=True)
df_license_21 = df_license_21.set_index("age_group")
df_license_21 = df_license_21.drop(columns=["성별"])
```

```python
df_license_21
```

```python
def change_index(index):
    if index == "16 - 19":
        return 10
    elif index == "20 - 24" or index == "25 - 29":
        return 20
    elif index == "30 - 34" or index == "35 - 39":
        return 30
    elif index == "40 - 44" or index == "45 - 49":
        return 40
    elif index == "50 - 54" or index == "55 - 59":
        return 50
    elif index == "60 - 64":
        return 60
    else:
        return 65
```

```python
df_license_21.index = df_license_21.index.map(change_index)
```

```python
df_license_21
```

```python
df_license_21["total"] = df_license_21.sum(axis=1)
df_license_21 = df_license_21[["total"]]
```

```python
df_license_21
```

```python
df_license_21.to_csv("./data/pre_driver_license_21.csv", encoding="utf-8")
```
