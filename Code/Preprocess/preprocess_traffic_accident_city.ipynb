```python
import pandas as pd
import numpy as np
import csv
```

```python
df_license_city = pd.read_excel("traffic_accident_city.xls", engine="xlrd")
```

```python
df_license_city = df_license_city.iloc[:-1]
df_license_city.rename(columns={"시도": "city"}, inplace=True)
df_license_city = df_license_city.set_index("city")
df_license_city = df_license_city.iloc[1:]
```

```python
df_license_city["10"] = (df_license_city["2019"] + df_license_city["2020"]  + df_license_city["2021"] + df_license_city["2022"] + df_license_city["2023"]) / 5
df_license_city = df_license_city.drop(columns=["2019", "2020", "2021", "2022", "2023"])
df_license_city["20"] = (df_license_city["2019.1"] + df_license_city["2020.1"]  + df_license_city["2021.1"] + df_license_city["2022.1"] + df_license_city["2023.1"]) / 5
df_license_city = df_license_city.drop(columns=["2019.1", "2020.1", "2021.1", "2022.1", "2023.1"])
df_license_city["30"] = (df_license_city["2019.2"] + df_license_city["2020.2"]  + df_license_city["2021.2"] + df_license_city["2022.2"] + df_license_city["2023.2"]) / 5
df_license_city = df_license_city.drop(columns=["2019.2", "2020.2", "2021.2", "2022.2", "2023.2"])
df_license_city["40"] = (df_license_city["2019.3"] + df_license_city["2020.3"]  + df_license_city["2021.3"] + df_license_city["2022.3"] + df_license_city["2023.3"]) / 5
df_license_city = df_license_city.drop(columns=["2019.3", "2020.3", "2021.3", "2022.3", "2023.3"])
df_license_city["50"] = (df_license_city["2019.4"] + df_license_city["2020.4"]  + df_license_city["2021.4"] + df_license_city["2022.4"] + df_license_city["2023.4"]) / 5
df_license_city = df_license_city.drop(columns=["2019.4", "2020.4", "2021.4", "2022.4", "2023.4"])
df_license_city["60"] = (df_license_city["2019.5"] + df_license_city["2020.5"]  + df_license_city["2021.5"] + df_license_city["2022.5"] + df_license_city["2023.5"]) / 5
df_license_city = df_license_city.drop(columns=["2019.5", "2020.5", "2021.5", "2022.5", "2023.5"])
df_license_city["65"] = (df_license_city["2019.6"] + df_license_city["2020.6"]  + df_license_city["2021.6"] + df_license_city["2022.6"] + df_license_city["2023.6"]) / 5
df_license_city = df_license_city.drop(columns=["2019.6", "2020.6", "2021.6", "2022.6", "2023.6"])
df_license_city = df_license_city.drop(columns=["2019.7", "2020.7", "2021.7", "2022.7", "2023.7"])
```

```python
df_license_city = df_license_city.transpose()
df_license_city.index.name = "age_group"
```
