```python
import pandas as pd
import numpy as np
import csv
```

```python
df_elderly = pd.read_csv("./data/elderly_people.csv", encoding="cp949")
```

```python
df_elderly.rename(columns={"행정구역": "city"}, inplace=True)
df_elderly["city"] = df_elderly["city"].replace({"충청북도  (4300000000)": "충북", "충청남도  (4400000000)": "충남", "전라북도  (4500000000)": "전북", "전라남도  (4600000000)": "전남", "경상북도  (4700000000)": "경북", "경상남도  (4800000000)": "경남"})
df_elderly["city"] = df_elderly["city"].str[:2]
df_elderly = df_elderly.set_index("city")
```

```python
df_elderly
```

```python
filtered_columns = [col for col in df_elderly.columns if "전체" in col]
df_elderly = df_elderly[filtered_columns]
```

```python
df_elderly
```

```python
for col in df_elderly.columns:
    if df_elderly[col].dtype == "object":
        df_elderly[col] = df_elderly[col].str.replace(",", "")
        df_elderly[col] = pd.to_numeric(df_elderly[col], errors='coerce')
```

```python
df_elderly
```

```python
df_elderly["total_2019"] = df_elderly.filter(regex=r"2019년\d{2}월_전체").sum(axis=1)
df_elderly["total_elderly_2019"] = df_elderly.filter(regex=r"2019년\d{2}월_65세이상전체").sum(axis=1)
df_elderly["total_2020"] = df_elderly.filter(regex=r"2020년\d{2}월_전체").sum(axis=1)
df_elderly["total_elderly_2020"] = df_elderly.filter(regex=r"2020년\d{2}월_65세이상전체").sum(axis=1)
df_elderly["total_2021"] = df_elderly.filter(regex=r"2021년\d{2}월_전체").sum(axis=1)
df_elderly["total_elderly_2021"] = df_elderly.filter(regex=r"2021년\d{2}월_65세이상전체").sum(axis=1)
df_elderly["total_2022"] = df_elderly.filter(regex=r"2022년\d{2}월_전체").sum(axis=1)
df_elderly["total_elderly_2022"] = df_elderly.filter(regex=r"2022년\d{2}월_65세이상전체").sum(axis=1)
df_elderly["total_2023"] = df_elderly.filter(regex=r"2023년\d{2}월_전체").sum(axis=1)
df_elderly["total_elderly_2023"] = df_elderly.filter(regex=r"2023년\d{2}월_65세이상전체").sum(axis=1)
df_elderly_5 = df_elderly[["total_2019", "total_elderly_2019", "total_2020", "total_elderly_2020", "total_2021", "total_elderly_2021", "total_2022", "total_elderly_2022", "total_2023", "total_elderly_2023"]]
```
