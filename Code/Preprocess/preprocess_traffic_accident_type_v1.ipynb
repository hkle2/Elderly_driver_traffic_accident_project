```python
import pandas as pd
import numpy as np
import csv
```

```python
df_traffic_type_5 = pd.read_excel("./data/traffic_accident_type_5.xls", engine="xlrd")
```

```python
df_traffic_type_5 = df_traffic_type_5.iloc[:-2]
df_traffic_type_5 = df_traffic_type_5.replace("-", np.nan)
df_traffic_type_5 = df_traffic_type_5.fillna(0)
```

```python
df_traffic_type_5
```

```python
columns = df_traffic_type_5.iloc[0]
df_traffic_type_5.columns = columns
df_traffic_type_5 = df_traffic_type_5[2:]
df_traffic_type_5 = df_traffic_type_5.groupby(df_traffic_type_5.columns, axis=1).sum()
```

```python
df_traffic_type_5
```

```python
df_traffic_type_5.rename(columns={"기본계획 연령별1": "age_group"}, inplace=True)
df_traffic_type_5 = df_traffic_type_5.drop(columns=["합계"])
```

```python
df_traffic_type_5
```

```python
df_traffic_type_5.columns = df_traffic_type_5.columns.str.replace(r"\([^)]*\)", "", regex=True).str.strip()
```

```python
df_traffic_type_5
```

```python
df_traffic_type_5["age_group"] = df_traffic_type_5["age_group"].replace({"20세이하": "10", "21~30세": "20", "31~40세": "30", "41~50세": "40", "51~60세": "50", "61~64세": "60", "65세이상": "65"})
```

```python
df_traffic_type_5
```

```python
df_traffic_type_5 = df_traffic_type_5.set_index("age_group")
```

```python
df_traffic_type_5
```

```python
df_traffic_type_5.to_csv("./data/pre_traffic_accident_type_5.csv", encoding="utf-8")
```
