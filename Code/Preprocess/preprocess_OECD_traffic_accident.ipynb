```python
import pandas as pd
import numpy as np
import csv
```

```python
df_OECD = pd.read_excel("OECD_traffic_accident.xls", engine="xlrd")
df_OECD = df_OECD.drop(df_OECD.index[-1])
```

```python
df_OECD = df_OECD.replace("-", np.nan)
df_OECD = df_OECD.fillna(0)
```

```python
df_OECD.rename(columns={"국가": "country"}, inplace=True)
year = ["2019.4", "2019.5", "2020.4", "2020.5", "2021.4", "2021.5"]
col = ["country"] + year
df_OECD = df_OECD[col]
```

```python
df_OECD.rename(columns={"2019.4": "death_population_2019", "2019.5": "death_car_2019", "2020.4": "death_population_2020", "2020.5": "death_car_2020", "2021.4": "death_population_2021", "2021.5": "death_car_2021"}, inplace=True)
df_OECD = df_OECD.drop(0)
df_OECD = df_OECD.set_index("country")
```
