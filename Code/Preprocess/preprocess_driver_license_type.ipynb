```python
import pandas as pd
import numpy as np
import csv
```

```python
df_license_type_5 = pd.read_excel("driver_license_type.xls", engine="xlrd")
```

```python
df_license_type_5 = df_license_type_5.iloc[1:-4]
df_license_type_5.rename(columns={"기본계획 연령별1": "age_group", "기준년도": "accident_type"}, inplace=True)
df_license_type_5 = df_license_type_5.replace("-", np.nan)
df_license_type_5 = df_license_type_5.fillna(0)
```

```python
df_license_type_5["total"] = (df_license_type_5["2018"] + df_license_type_5["2019"] + df_license_type_5["2020"]  + df_license_type_5["2021"] + df_license_type_5["2022"]) / 5
df_license_type_5 = df_license_type_5.drop(columns=["2018", "2019", "2020", "2021", "2022"])
df_license_type_5["제1종"] = (df_license_type_5["2018.1"] + df_license_type_5["2019.1"] + df_license_type_5["2020.1"]  + df_license_type_5["2021.1"] + df_license_type_5["2022.1"]) / 5
df_license_type_5 = df_license_type_5.drop(columns=["2018.1", "2019.1", "2020.1", "2021.1", "2022.1"])
df_license_type_5["제2종"] = (df_license_type_5["2018.2"] + df_license_type_5["2019.2"] + df_license_type_5["2020.2"]  + df_license_type_5["2021.2"] + df_license_type_5["2022.2"]) / 5
df_license_type_5 = df_license_type_5.drop(columns=["2018.2", "2019.2", "2020.2", "2021.2", "2022.2"])
df_license_type_5["연습면허"] = (df_license_type_5["2018.3"] + df_license_type_5["2019.3"] + df_license_type_5["2020.3"]  + df_license_type_5["2021.3"] + df_license_type_5["2022.3"]) / 5
df_license_type_5 = df_license_type_5.drop(columns=["2018.3", "2019.3", "2020.3", "2021.3", "2022.3"])
df_license_type_5["무면허"] = (df_license_type_5["2018.4"] + df_license_type_5["2019.4"] + df_license_type_5["2020.4"]  + df_license_type_5["2021.4"] + df_license_type_5["2022.4"]) / 5
df_license_type_5 = df_license_type_5.drop(columns=["2018.4", "2019.4", "2020.4", "2021.4", "2022.4"])
df_license_type_5["국제면허"] = (df_license_type_5["2018.5"] + df_license_type_5["2019.5"] + df_license_type_5["2020.5"]  + df_license_type_5["2021.5"] + df_license_type_5["2022.5"]) / 5
df_license_type_5 = df_license_type_5.drop(columns=["2018.5", "2019.5", "2020.5", "2021.5", "2022.5"])
df_license_type_5["기타"] = ((df_license_type_5["2018.6"] + df_license_type_5["2019.6"] + df_license_type_5["2020.6"]  + df_license_type_5["2021.6"] + df_license_type_5["2022.6"]) / 5) + ((df_license_type_5["2018.7"] + df_license_type_5["2019.7"] + df_license_type_5["2020.7"]  + df_license_type_5["2021.7"] + df_license_type_5["2022.7"]) / 5)
df_license_type_5 = df_license_type_5.drop(columns=["2018.6", "2019.6", "2020.6", "2021.6", "2022.6", "2018.7", "2019.7", "2020.7", "2021.7", "2022.7"])
```

```python
df_license_type_5["age_group"] = df_license_type_5["age_group"].replace({"합계": "all_age", "20세이하": 10, "21~30세": 20, "31~40세": 30, "41~50세": 40, "51~60세": 50, "61~64세": 60, "65세이상": 65})
df_license_type_5["accident_type"] = df_license_type_5["accident_type"].replace({"사고[건]": "accident", "사망[명]": "dead", "부상[명]": "injured"})
df_license_type_5 = df_license_type_5.set_index("age_group")
```
