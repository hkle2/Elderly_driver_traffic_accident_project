```python
import pandas as pd
import numpy as np
import csv
```

```python
df_license_time = pd.read_excel("traffic_accident_time.xls", engine="xlrd")
```

```python
df_license_time = df_license_time.iloc[1:-4]
df_license_time.rename(columns={"기본계획 연령별1": "age_group", "기준년도": "accident_type"}, inplace=True)
```

```python
df_license_time["total"] = (df_license_time["2019"] + df_license_time["2020"]  + df_license_time["2021"] + df_license_time["2022"] + df_license_time["2023"]) / 5
df_license_time = df_license_time.drop(columns=["2019", "2020", "2021", "2022", "2023"])
df_license_time["time_0_2"] = (df_license_time["2019.1"] + df_license_time["2020.1"]  + df_license_time["2021.1"] + df_license_time["2022.1"] + df_license_time["2023.1"]) / 5
df_license_time = df_license_time.drop(columns=["2019.1", "2020.1", "2021.1", "2022.1", "2023.1"])
df_license_time["time_2_4"] = (df_license_time["2019.2"] + df_license_time["2020.2"]  + df_license_time["2021.2"] + df_license_time["2022.2"] + df_license_time["2023.2"]) / 5
df_license_time = df_license_time.drop(columns=["2019.2", "2020.2", "2021.2", "2022.2", "2023.2"])
df_license_time["time_4_6"] = (df_license_time["2019.3"] + df_license_time["2020.3"]  + df_license_time["2021.3"] + df_license_time["2022.3"] + df_license_time["2023.3"]) / 5
df_license_time = df_license_time.drop(columns=["2019.3", "2020.3", "2021.3", "2022.3", "2023.3"])
df_license_time["time_6_8"] = (df_license_time["2019.4"] + df_license_time["2020.4"]  + df_license_time["2021.4"] + df_license_time["2022.4"] + df_license_time["2023.4"]) / 5
df_license_time = df_license_time.drop(columns=["2019.4", "2020.4", "2021.4", "2022.4", "2023.4"])
df_license_time["time_8_10"] = (df_license_time["2019.5"] + df_license_time["2020.5"]  + df_license_time["2021.5"] + df_license_time["2022.5"] + df_license_time["2023.5"]) / 5
df_license_time = df_license_time.drop(columns=["2019.5", "2020.5", "2021.5", "2022.5", "2023.5"])
df_license_time["time_10_12"] = (df_license_time["2019.6"] + df_license_time["2020.6"]  + df_license_time["2021.6"] + df_license_time["2022.6"] + df_license_time["2023.6"]) / 5
df_license_time = df_license_time.drop(columns=["2019.6", "2020.6", "2021.6", "2022.6", "2023.6"])
df_license_time["time_12_14"] = (df_license_time["2019.7"] + df_license_time["2020.7"]  + df_license_time["2021.7"] + df_license_time["2022.7"] + df_license_time["2023.7"]) / 5
df_license_time = df_license_time.drop(columns=["2019.7", "2020.7", "2021.7", "2022.7", "2023.7"])
df_license_time["time_14_16"] = (df_license_time["2019.8"] + df_license_time["2020.8"]  + df_license_time["2021.8"] + df_license_time["2022.8"] + df_license_time["2023.8"]) / 5
df_license_time = df_license_time.drop(columns=["2019.8", "2020.8", "2021.8", "2022.8", "2023.8"])
df_license_time["time_16_18"] = (df_license_time["2019.9"] + df_license_time["2020.9"]  + df_license_time["2021.9"] + df_license_time["2022.9"] + df_license_time["2023.9"]) / 5
df_license_time = df_license_time.drop(columns=["2019.9", "2020.9", "2021.9", "2022.9", "2023.9"])
df_license_time["time_18_20"] = (df_license_time["2019.10"] + df_license_time["2020.10"]  + df_license_time["2021.10"] + df_license_time["2022.10"] + df_license_time["2023.10"]) / 5
df_license_time = df_license_time.drop(columns=["2019.10", "2020.10", "2021.10", "2022.10", "2023.10"])
df_license_time["time_20_22"] = (df_license_time["2019.11"] + df_license_time["2020.11"]  + df_license_time["2021.11"] + df_license_time["2022.11"] + df_license_time["2023.11"]) / 5
df_license_time = df_license_time.drop(columns=["2019.11", "2020.11", "2021.11", "2022.11", "2023.11"])
df_license_time["time_22_24"] = (df_license_time["2019.12"] + df_license_time["2020.12"]  + df_license_time["2021.12"] + df_license_time["2022.12"] + df_license_time["2023.12"]) / 5
df_license_time = df_license_time.drop(columns=["2019.12", "2020.12", "2021.12", "2022.12", "2023.12"])
```

```python
df_license_time["age_group"] = df_license_time["age_group"].replace({"합계": "all_age", "20세이하": 10, "21~30세": 20, "31~40세": 30, "41~50세": 40, "51~60세": 50, "61~64세": 60, "65세이상": 65})
df_license_time["accident_type"] = df_license_time["accident_type"].replace({"사고[건]": "accident", "사망[명]": "dead", "부상[명]": "injured"})
df_license_time = df_license_time.set_index("age_group")
```
