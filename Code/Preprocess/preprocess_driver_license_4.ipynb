# 고령자 교통사고

```python
import pandas as pd
import csv
```

```python
df_license_4 = pd.read_csv("./data/driver_license_4.csv", encoding="utf-8")
```

```python
df_license_4 = df_license_4.iloc[3:]
year = ["2019", "2020", "2021", "2022"]
col = ["연령대별(1)"] + year
df_license_4 = df_license_4[col]
```

```python
df_license_4
```

```python
df_license_4.rename(columns={"연령대별(1)": "age_group", "2019": "total_2019", "2020": "total_2020", "2021": "total_2021", "2022": "total_2022"}, inplace=True)
df_license_4 = df_license_4.set_index("age_group")
```

```python
df_license_4
```

```python
def change_index(index):
    if int(index[0]) == 1:
        return 10
    elif int(index[0]) == 2:
        return 20
    elif int(index[0]) == 3:
        return 30
    elif int(index[0]) == 4:
        return 40
    elif int(index[0]) == 5:
        return 50
    elif 60 <= int(index[:2]) <= 64:
        return 60
    else:
        return 65
```

```python
df_license_4.index = df_license_4.index.map(change_index)
```

```python
df_license_4
```

```python
df_license_4.to_csv("./data/pre_driver_license_4.csv", encoding="utf-8")
```
