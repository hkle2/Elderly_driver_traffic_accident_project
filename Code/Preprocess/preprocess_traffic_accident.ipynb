# 고령자 교통사고

```python
import pandas as pd
import csv
```

```python
df_traffic_accident = pd.read_csv("./data/traffic_accident.csv", encoding="cp949")
```

```python
df_traffic_accident = df_traffic_accident[df_traffic_accident["가해자연령층별(1)"] != '불명']
year = ["2018", "2018.1", "2018.2", "2019", "2019.1", "2019.2", "2020", "2020.1", "2020.2", "2021", "2021.1", "2021.2", "2022", "2022.1", "2022.2"]
col = ["가해자연령층별(1)"] + ["월별(1)"] + year
df_traffic_accident = df_traffic_accident[col]
```

```python
df_traffic_accident
```

```python
df_traffic_accident.rename(columns={"가해자연령층별(1)": "age_group", "월별(1)": "month", "2018": "total_2018", "2018.1": "dead_2018", "2018.2": "injured_2018", "2019": "total_2019", "2019.1": "dead_2019", "2019.2": "injured_2019", "2020": "total_2020", "2020.1": "dead_2020", "2020.2": "injured_2020", "2021": "total_2021", "2021.1": "dead_2021", "2021.2": "injured_2021", "2022": "total_2022", "2022.1": "dead_2022", "2022.2": "injured_2022"}, inplace=True)
df_traffic_accident = df_traffic_accident.drop(0)
df_traffic_accident = df_traffic_accident[df_traffic_accident["month"] == "전체"]
df_traffic_accident = df_traffic_accident.drop(columns=["month"])
```

```python
df_traffic_accident
```

```python
df_traffic_accident["age_group"] = df_traffic_accident["age_group"].replace({"20세이하": 10, "21-30세": 20, "31-40세": 30, "41-50세": 40, "51-60세": 50, "61-64세": 60, "65-70세": 65, "71세이상": 65})
df_traffic_accident = df_traffic_accident.set_index("age_group")
```

```python
df_traffic_accident
```

```python
df_traffic_accident.to_csv("./data/pre_traffic_accident.csv", encoding="utf-8")
```
